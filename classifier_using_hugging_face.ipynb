```python
from transformers import pipeline
```

```python
classifier=pipeline("zero-shot-classification")
```

```python
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

```python
classifier(
    "The forest was alive with the sound of chirping birds and rustling leaves.",
    candidate_labels=["nature", "technology", "sports"],
)
```

```python
classifier(
    "The team celebrated their victory with a dramatic final goal in overtime.",
    candidate_labels=["sports", "music", "politics"],
)
```

```python
classifier(
    "Daily meditation has been shown to reduce stress and improve focus.",
    candidate_labels=["health", "finance", "education"],
)
```
